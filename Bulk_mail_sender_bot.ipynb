```python
import smtplib
import os
from email.mime.multipart import MIMEMultipart
from email.mime.base import MIMEBase
from email.mime.text import MIMEText
from email import encoders
import mimetypes
prev="emp"
def login(email,password):
  try:
   # if(prev==server):
    #  server.close()
    server=smtplib.SMTP('smtp.gmail.com',587)
    server.starttls()
    server.login(email,password)
    prev=server
    return server
  except:
    print("check login Information error in accessing")

def email_sender(Esender,Ercv,content,server):
  try:
    server.sendmail(Esender,Ercv,content)
  except:
    print("Error in sending Email")
def thread(message):
  with open('login.txt','r')as Login:
    for sender in Login:
      Email= sender[:sender.index(':')]
      password= sender[sender.index(':'):]
      password=password.removeprefix(":")
      server=login(Email,password)
      with open('mail list.txt', 'r') as elist:
        for email in elist:
          email_sender(Email,email,message,server)
          print("Email sent to ",email,"  by ",Email)
with open("M_subject.txt",'r')as sub:
  subject=sub.read()
with open('Message_Token.txt', 'r') as M_T:
    message = M_T.read()
msg = MIMEMultipart()
msg["Subject"] = subject
# Add the message body
msg.attach(MIMEText(message, "plain"))

path = "Attachments"
contents = os.listdir(path)
for names in contents:
  mime_type, encoding = mimetypes.guess_type(f"Attachments/{names}")
  with open(f"Attachments/{names}", "rb") as attachment:
      part = MIMEBase("application", "octet-stream")
      part.set_payload(attachment.read())
      encoders.encode_base64(part)
      part.add_header("Content-Disposition", f"attachment; filename= {names}")
      part.add_header("Content-Type", f"{mime_type}") # set the content type of the attachment
      msg.attach(part)
text = msg.as_string()
thread(text)
```

```python
import imaplib
import email

# Login credentials
username = ''
password = ''

# Connect to Gmail's IMAP server
mail = imaplib.IMAP4_SSL('imap.gmail.com')
mail.login(username, password)
mail.select('inbox')

# Fetch the latest email
status, data = mail.search(None, 'ALL')
latest_email_id = data[0].split()[-1]
status, data = mail.fetch(latest_email_id, '(RFC822)')
raw_email = data[0][1]

# Parse the email
email_message = email.message_from_bytes(raw_email)
print('Subject:', email_message['Subject'])
print('From:', email_message['From'])
print('To:', email_message['To'])
print(email_message.decode(''))
```
